Load the sonnets from the file provided. `sonnets` is indexed by sonnet number (0-indexed, so shifted -1 from sonnet number), then by line number (again, 0-indexed).

`sonnets[0][3]` for example returns the 4th line of the first sonnet.

```python
with open('data/shakespeare.txt') as f:
    lines = [line.lower().strip('\n ') for line in f]
sonnets = []
ln_start = 0
ln_end = 0
for ln, content in enumerate(lines):
    if content.isdigit():
        ln_start = ln + 1
    elif not content:
        if ln - 1 == ln_end:
            sonnets.append(lines[ln_start:ln_end + 1])
    elif ln + 1 == len(lines):
        sonnets.append(lines[ln_start:ln + 1])
    else:
        ln_end = ln
```

Generate the line ending punctuations, to regenerate them later

```python
sonnet_ends = []
end_chars = set()
for s in sonnets:
    line_ends = []
    for l in s:
        if l[-1].isalpha():
            line_ends.append('')
        else:
            line_ends.append(l[-1])
    sonnet_ends.append(line_ends)
    end_chars.update(line_ends)
    
end_chars = list(sorted(end_chars))
end_char_indices = dict((c, i) for i, c in enumerate(end_chars))
indices_end_char = dict((i, c) for i, c in enumerate(end_chars))
```

Generate syllable map, word maps, and reverse word maps.

```python
from collections import defaultdict
words = []
syllable_dict = {}
rev_syllable_dict = defaultdict(list)
rev_end_syllable_dict = defaultdict(list)

with open('data/Syllable_dictionary.txt') as f:
    for i, line in enumerate(f):
        tokens = line.strip().split(' ')
        words.append(tokens[0])
        syllable_dict[i] = tokens[1:]
        for syl in tokens[1:]:
            if syl[0] == 'E':
                rev_end_syllable_dict[int(syl[1:])].append(i)
            else:
                rev_syllable_dict[int(syl)].append(i)
```

```python
word_indices = dict((c, i) for i, c in enumerate(words))
indices_word = dict((i, c) for i, c in enumerate(words))
```

```python
stress_start = []
unstress_start = []
stress_end = []
unstress_end = []
X = []
lens = []
for s in sonnets:
    smapped = []
    for l in s:
        syl_count = 0
        ws = l.split(' ')
        lens.append(len(ws))
        lmapped = []
        for w in ws:
            if w not in words:
                w = w.rstrip("!'(),.:;?")
                if w not in words:
                    w = w.strip("!'(),.:;?")
            wi = word_indices[w]
            lmapped.append(wi)
        X.extend(reversed(lmapped))
```

Train a HMM on the lines in the sonnets

```python
from hmmlearn.hmm import MultinomialHMM
```

```python
model = MultinomialHMM(n_components=50, n_iter=100, verbose=True)
model.fit(np.reshape(np.array(X), (-1, 1)), np.array(lens))
```

```python
A = model.transmat_
O = model.emissionprob_
A_start = model.startprob_
```

```python
gen_lines = [None for _ in range(14)]
n_states = A.shape[0]
n_words = 3205
for i, xxx in enumerate(np.random.choice(len(q_rhymes), 7)):
    if i == 6:
        line_f = 12
        line_s = 13
    else:
        line_f = int((i) / 2) * 4 + i % 2
        line_s = line_f + 2
    print(line_f, line_s)
    gen_line_f = []
    gen_line_s = []
    if not i == 6:
        print(q_rhymes[xxx])
        word_f, word_s = q_rhymes[xxx]
    if i == 6:
        word_f, word_s = c_rhymes[np.random.choice(len(c_rhymes))]
    
    ind_f = word_indices[word_f]
    ps = O[:, ind_f] * A_start
    ps /= np.sum(ps)
    start_state = np.random.choice(n_states, p=ps)
    gen_line_f += [word_f]
    syl_str = syllable_dict[ind_f][0]
    if syl_str[0] == 'E': syl_str = syl_str[1:]
    syl_count = int(syl_str)
    state = start_state
    while syl_count < 9 :
        state = np.random.choice(n_states, p=A[state])
        wi = np.random.choice(n_words, p=O[state])
        syl_str = syllable_dict[wi][-1]
        if syl_str[0] == 'E': syl_str = syl_str[1:]
        syl_count += int(syl_str)
        gen_line_f.append(indices_word[wi])
    gen_lines[line_f] = gen_line_f
    
    ind_s = word_indices[word_s]
    ps = O[:, ind_s] * A_start
    ps /= np.sum(ps)
    start_state = np.random.choice(n_states, p=ps)
    gen_line_s += [word_s]
    syl_str = syllable_dict[ind_s][0]
    if syl_str[0] == 'E': syl_str = syl_str[1:]
    syl_count = int(syl_str)
    state = start_state
    while syl_count < 9 :
        state = np.random.choice(n_states, p=A[state])
        wi = np.random.choice(n_words, p=O[state])
        syl_str = syllable_dict[wi][-1]
        if syl_str[0] == 'E': syl_str = syl_str[1:]
        syl_count += int(syl_str)
        gen_line_s.append(indices_word[wi])
    gen_lines[line_s] = gen_line_s
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def mask():
    # Parameters.
    r = 128
    d = 2 * r + 1

    # Get points in a circle.
    y, x = np.ogrid[-r:d-r, -r:d-r]
    circle = (x**2 + y**2 <= r**2)

    # Create mask.
    mask = 255 * np.ones((d, d), dtype=np.uint8)
    mask[circle] = 0

    return mask

def text_to_wordcloud(text, max_words=50, title='', show=True):
    plt.close('all')

    # Generate a wordcloud image.
    wordcloud = WordCloud(random_state=0,
                          max_words=max_words,
                          background_color='white',
                          mask=mask()).generate(text)

    # Show the image.
    if show:
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(title, fontsize=24)
        plt.savefig('wordclouds/' + title)
        plt.show()

    return wordcloud
```

```python
max_words=50
show=True
M = 1000
n_states = len(A)
n_em = O.shape[1]
wordclouds = []

# For each state, get a list of observations that have been emitted
# from that state.
obs_count = []
for i in range(n_states):
    obs_lst = np.random.choice(n_em, size=M, p=O[i])
    obs_count.append(obs_lst)

# For each state, convert it into a wordcloud.
for i in range(n_states):
    obs_lst = obs_count[i]
    sentence = [indices_word[j] for j in obs_lst]
    sentence_str = ' '.join(sentence)

    wordclouds.append(text_to_wordcloud(sentence_str, max_words=max_words, title='State %d' % i, show=show))
```
